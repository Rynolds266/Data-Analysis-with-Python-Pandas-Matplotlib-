```python
#2-DATA COLLECTION AND LOADING

import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns',2000)
pd.set_option('display.width',1000)
pd.set_option('display.float_format', '{:.2f}'.format)



file_source=r'C:\Users\decym\Downloads\owid-co2-data.csv'

df=pd.read_csv(file_source)

print(df.head())
```

```python
                                  #Cleaning Data:
    
#like you can notice to our previous cell there are missing values for critical columns like co2 and population
# 'co2_growth_prct', 'share_global_co2', 'share_global_cumulative_co2', 'gdp', 'co2_per_gdp'

# I have decided to drop some colums that contain huge missing values

columns_to_keep=[
     'country', 'year', 'co2', 'population', 'co2_per_capita', 'cumulative_co2',
    'coal_co2', 'oil_co2', 'gas_co2', 'cement_co2'
   
]

cleaned_data=df[columns_to_keep]

print(cleaned_data.head(10))
```

```python
#3. Data Cleaning and Preparation
#In this project, the dataset contains missing values in several columns. 
#To ensure data integrity and maintain meaningful analysis, we applied the following strategies:
#d we dropped  columns with more than 80% missing values.
#
#
#


cleaned_data.loc[0 ,'oil_co2']=0.07

cleaned_data['oil_co2'].fillna(cleaned_data['oil_co2'].mean(), inplace=False)
cleaned_data['gas_co2'].fillna(cleaned_data['gas_co2'].mean(), inplace=False)

for x in cleaned_data.index:
   if pd.isnull(cleaned_data.loc[x,'cement_co2']):
    
      cleaned_data.loc[x ,'cement_co2']=cleaned_data['cement_co2'].mean()
    
   if  pd.isnull(cleaned_data.loc[x,'co2']):
          cleaned_data.loc[x ,'co2']=cleaned_data['cement_co2'].median()
   

        

#checking for missing value in our dataset



print(cleaned_data)
```

```python
#handle missimg values


#cleaned_data['co2']=cleaned_data['co2'].fillna(cleaned_data['co2'].median(), inplace=False)

#cleaned_data['population'] = cleaned_data['population'].interpolate()

#cleaned_data['co2_per_capita'] = cleaned_data['co2'] / cleaned_data['population']

#cleaned_data['cumulative_co2'] = cleaned_data.groupby('country')['co2'].cumsum()

#cleaned_data['coal_co2'].fillna(cleaned_data['coal_co2'].median(), inplace=False)

#cleaned_data['gas_co2'].fillna(cleaned_data['gas_co2'].median(), inplace=False)

#cleaned_data['cement_co2'].fillna(cleaned_data['cement_co2'].median(), inplace=False)


cleaned_data.loc[:, 'co2'] = cleaned_data['co2'].fillna(cleaned_data['co2'].median())
cleaned_data.loc[:, 'population'] = cleaned_data['population'].interpolate()
cleaned_data.loc[:, 'co2_per_capita'] = cleaned_data['co2'] / cleaned_data['population']
cleaned_data.loc[:, 'cumulative_co2'] = cleaned_data.groupby('country')['co2'].cumsum()
cleaned_data.loc[:, 'year'] = pd.to_datetime(cleaned_data['year'], format='%Y')
cleaned_data.loc[:, 'cumulative_country_co2'] = cleaned_data.groupby('country')['co2'].cumsum()


cleaned_data.loc[:, 'year'] = pd.to_datetime(cleaned_data['year'], format='%Y')
cleaned_data.loc[:, 'cumulative_country_co2'] = cleaned_data.groupby('country')['co2'].cumsum()




#convert date to datetime

#cleaned_data['year']=pd.to_datetime(cleaned_data['year'], format='%Y')

#CREATE NEW FEATURE 
#Calculate cumulative CO2 emissions over the years for each country.
#cleaned_data['cumulative_country_co2'] = cleaned_data.groupby('country')['co2'].cumsum()


print(cleaned_data)

```

```python
missvalue=cleaned_data.isnull().sum()
verify_missing=missvalue[missvalue>0]
print(verify_missing)
```

```python


#4. Exploratory Data Analysis (EDA)


# Summary statistics for key numerical columns

print(cleaned_data[['co2', 'population', 'co2_per_capita', 'cumulative_co2']].describe())


# Median
print(cleaned_data[['co2', 'population','co2_per_capita']].median())


# Min and Max values
print("Minimum CO2 Emissions: ", cleaned_data['co2'].min())
print("Maximum CO2 Emissions: ", cleaned_data['co2'].max())


```

```python
# Group by country and sum total CO2 emission


filtered_data = cleaned_data[
    ~cleaned_data['country'].str.contains(
        'World|Asia|Europe|income|Union|America|Africa|Oceania', 
        case=False, na=False
    )
]


top_5_countries = (
    filtered_data.groupby('country')['co2']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .index
)

top_5_data = filtered_data[filtered_data['country'].isin(top_5_countries)]


plt.figure(figsize=(12, 7))
for country in top_5_countries:
    country_data = top_5_data[top_5_data['country'] == country]
    plt.plot(country_data['year'], country_data['co2'], label=country)


plt.title('CO2 Emissions Over Time for Top 5 Emitting Countries')
plt.xlabel('Year')
plt.ylabel('Total CO2 Emissions (in million tons)')
plt.legend(title='Country')
plt.grid(True)
plt.show()


```
